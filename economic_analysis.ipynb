## Aim
To analyse how economic factors such as unemployment and cost of living influence mental health outcomes.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("economic_data.csv.csv")
df = df.iloc[::-1]

plt.plot(df["Period"], df["Unemployment"], marker="o", label="Unemployment Rate (%)")
plt.plot(df["Period"], df["Anxiety "], marker="o", label="Anxiety Mean (0-10)")

plt.title("Unemployment Rate and Anxiety Levels Over Time")
plt.xlabel("Period")
plt.ylabel("Value")

plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()

plt.show()
```

## Insight

Unemployment rates show a general downward trend over time, while anxiety levels remain relatively stable with slight fluctuations.

Despite improvements in employment conditions, anxiety levels do not decrease proportionally, suggesting that factors beyond employment status may contribute to mental health outcomes.

However, periods of higher unemployment coincide with slightly elevated anxiety levels, indicating that economic instability may still play a role in shaping psychological wellbeing.

This highlights the complexity of the relationship between economic conditions and mental health.
